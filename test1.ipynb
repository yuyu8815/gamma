```python
def get_spot(instr, d):
    try:
        df = cmd.book_printer_v2(f'{instr}', d,source='SRC_TWSE.ORIG_CPBQ.DEST_CPBQ').between_time('09:00','13:31').resample('1T').last().fillna(method='ffill')
        df['ask_px'] = np.where(df['ask_px'].values == 0.0, df['bid_px'], df['ask_px'])
        df['bid_px'] = np.where(df['bid_px'].values == 0.0, df['ask_px'], df['bid_px'])
        df['mkt_mid'] = (df['bid_px'] + df['ask_px']) / 2
        df = df[['mkt_mid']]
        df.columns = [instr]
        return df
    except Exception as e:
        print(f"error {d} {instr}: {e}")
        return None
```
