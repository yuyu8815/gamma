```python
import numpy as np
from scipy.stats import norm

def vega_black76(F, K, T, sigma):
    d1 = (np.log(F / K) + 0.5 * sigma * sigma * T) / (sigma * np.sqrt(T))
    vega = np.exp(0) * F * norm.pdf(d1) * np.sqrt(T)  # r=0 for TXO
    return vega/100
```
