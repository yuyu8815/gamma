```python
import re

def classify_exp(exp_str: str) -> str | None:
    # weekly like 202506w1, 202506w2 ...
    m = re.match(r"(\d{6})w([1-5])$", exp_str)
    if m:
        week_idx = int(m.group(2))
        return f"w{week_idx}"
    # monthly like 202506, 202507 ...
    if re.match(r"^\d{6}$", exp_str):
        # TODO: here you should map to f1/f2/f3/f4/f5/spot
        # Example: just treat the nearest monthly as f1, next as f2, etc.
        # This part depends on your own definition.
        return None
    return None

df["exp_group"] = df["exp"].astype(str).map(classify_exp)
```

```python
import pandas as pd

# 假設你的 df 長這樣，且已經有 exp_group 欄位
# columns: time, symid, exp, ref_vol, ..., exp_group

# 1. 建出日期欄位
df["date"] = df["time"].dt.date

# 2. 定義優先順序
priority_order = ["w1", "f1", "w2", "f2", "spot", "f3", "w4", "f4", "w5", "f5"]
priority_map = {k: i for i, k in enumerate(priority_order)}

# 3. 把 exp_group 映射成優先分數，沒在列表內的丟到很後面
df["priority"] = df["exp_group"].map(priority_map).fillna(999).astype(int)

# 4. 先壓到「date + exp」層級，找出每個 exp 在當天的最小優先值
daily_exp_priority = (
    df.groupby(["date", "exp"], as_index=False)["priority"]
      .min()
)

# 5. 對每一天挑出優先值最小的那個 exp
best_exp_per_day = (
    daily_exp_priority
    .sort_values(["date", "priority"])
    .drop_duplicates("date")  # 每個 date 留第一個
    .rename(columns={"exp": "chosen_exp"})
    [["date", "chosen_exp"]]
)

# 6. 把每日選出的 exp merge 回去，得到 df_spot
df_spot = (
    df.merge(
        best_exp_per_day,
        left_on=["date", "exp"],
        right_on=["date", "chosen_exp"],
        how="inner",
    )
    .drop(columns=["chosen_exp"])
)

# df_spot: 每一天只有一個 exp，但保留那一天所有時間點的資料
```
